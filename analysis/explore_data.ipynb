```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import seaborn as sns
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Set display options for better readability
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)

# Define data directory
DATA_DIR = Path("./")

print("="*80)
print("TENNIS DATA EXPLORATORY ANALYSIS")
print("="*80)

# ============================================================================
# SECTION 1: LOAD ALL DATASETS
# ============================================================================

print("\n" + "="*80)
print("SECTION 1: LOADING DATASETS")
print("="*80)

# Load players
try:
    df_players = pd.read_csv(DATA_DIR / "atp_players.csv")
    print(f"âœ“ Loaded atp_players.csv: {len(df_players):,} rows")
except Exception as e:
    print(f"âœ— Error loading players: {e}")
    df_players = None

# Load current rankings
try:
    df_rankings_current = pd.read_csv(DATA_DIR / "atp_rankings_current.csv")
    print(f"âœ“ Loaded atp_rankings_current.csv: {len(df_rankings_current):,} rows")
except Exception as e:
    print(f"âœ— Error loading current rankings: {e}")
    df_rankings_current = None

# Load historical rankings
try:
    df_rankings_20s = pd.read_csv(DATA_DIR / "atp_rankings_20s.csv")
    print(f"âœ“ Loaded atp_rankings_20s.csv: {len(df_rankings_20s):,} rows")
except Exception as e:
    print(f"âœ— Error loading 20s rankings: {e}")
    df_rankings_20s = None

# Load 2024 matches
try:
    df_matches_2024 = pd.read_csv(DATA_DIR / "atp_matches_2024.csv")
    print(f"âœ“ Loaded atp_matches_2024.csv: {len(df_matches_2024):,} rows")
except Exception as e:
    print(f"âœ— Error loading 2024 matches: {e}")
    df_matches_2024 = None

# Load 2023 matches (if available)
try:
    df_matches_2023 = pd.read_csv(DATA_DIR / "atp_matches_2023.csv")
    print(f"âœ“ Loaded atp_matches_2023.csv: {len(df_matches_2023):,} rows")
except Exception as e:
    print(f"âœ— Error loading 2023 matches: {e}")
    df_matches_2023 = None

# ============================================================================
# SECTION 2: PLAYERS TABLE ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("SECTION 2: PLAYERS TABLE ANALYSIS")
print("="*80)

if df_players is not None:
    print("\n--- BASIC INFO ---")
    print(f"Total Players: {len(df_players):,}")
    print(f"Total Columns: {len(df_players.columns)}")
    print(f"\nColumn Names:\n{list(df_players.columns)}")
    
    print("\n--- SAMPLE DATA (First 5 rows) ---")
    print(df_players.head())
    
    print("\n--- DATA TYPES ---")
    print(df_players.dtypes)
    
    print("\n--- MISSING VALUES ---")
    missing = df_players.isnull().sum()
    missing_pct = (missing / len(df_players) * 100).round(2)
    missing_df = pd.DataFrame({
        'Missing_Count': missing,
        'Missing_Percent': missing_pct
    })
    print(missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False))
    
    print("\n--- BASIC STATISTICS ---")
    print(df_players.describe(include='all'))
    
    print("\n--- POTENTIAL PRIMARY KEY ANALYSIS ---")
    # Check if player_id is unique
    if 'player_id' in df_players.columns:
        print(f"Total player_id: {len(df_players)}")
        print(f"Unique player_id: {df_players['player_id'].nunique()}")
        print(f"Duplicates: {len(df_players) - df_players['player_id'].nunique()}")
        
        if len(df_players) == df_players['player_id'].nunique():
            print("âœ“ player_id is UNIQUE - can be PRIMARY KEY")
        else:
            print("âœ— player_id has DUPLICATES - investigate further")
            duplicates = df_players[df_players.duplicated(subset=['player_id'], keep=False)]
            print(f"\nDuplicate player_ids:\n{duplicates.head(10)}")
    
    print("\n--- VALUE DISTRIBUTIONS FOR KEY COLUMNS ---")
    categorical_cols = df_players.select_dtypes(include=['object']).columns
    for col in categorical_cols[:5]:  # First 5 categorical columns
        print(f"\n{col} - Top 10 values:")
        print(df_players[col].value_counts().head(10))

# ============================================================================
# SECTION 3: RANKINGS TABLE ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("SECTION 3: RANKINGS TABLE ANALYSIS")
print("="*80)

if df_rankings_current is not None:
    print("\n--- CURRENT RANKINGS ---")
    print(f"Total Records: {len(df_rankings_current):,}")
    print(f"Columns: {list(df_rankings_current.columns)}")
    
    print("\n--- SAMPLE DATA (Top 10 ranked players) ---")
    print(df_rankings_current.head(10))
    
    print("\n--- MISSING VALUES ---")
    missing = df_rankings_current.isnull().sum()
    missing_pct = (missing / len(df_rankings_current) * 100).round(2)
    missing_df = pd.DataFrame({
        'Missing_Count': missing,
        'Missing_Percent': missing_pct
    })
    print(missing_df[missing_df['Missing_Count'] > 0])
    
    print("\n--- DATA TYPES ---")
    print(df_rankings_current.dtypes)

if df_rankings_20s is not None:
    print("\n--- HISTORICAL RANKINGS (2020s) ---")
    print(f"Total Records: {len(df_rankings_20s):,}")
    print(f"Columns: {list(df_rankings_20s.columns)}")
    
    # Check date column
    if 'ranking_date' in df_rankings_20s.columns:
        df_rankings_20s['ranking_date'] = pd.to_datetime(df_rankings_20s['ranking_date'])
        print(f"\nDate Range: {df_rankings_20s['ranking_date'].min()} to {df_rankings_20s['ranking_date'].max()}")
        print(f"Unique Dates: {df_rankings_20s['ranking_date'].nunique()}")
        print(f"Records per Date (avg): {len(df_rankings_20s) / df_rankings_20s['ranking_date'].nunique():.0f}")
    
    print("\n--- GRAIN ANALYSIS (What does one row represent?) ---")
    # Check if combination of date + player is unique
    if 'ranking_date' in df_rankings_20s.columns and 'player' in df_rankings_20s.columns:
        total_rows = len(df_rankings_20s)
        unique_combos = df_rankings_20s.groupby(['ranking_date', 'player']).size().count()
        print(f"Total rows: {total_rows:,}")
        print(f"Unique (date + player) combinations: {unique_combos:,}")
        if total_rows == unique_combos:
            print("âœ“ Grain is: ONE ranking per player per date")
        else:
            print("âœ— Duplicates exist - grain is unclear")
    
    print("\n--- SAMPLE DATA ---")
    print(df_rankings_20s.head(10))

# ============================================================================
# SECTION 4: MATCHES TABLE ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("SECTION 4: MATCHES TABLE ANALYSIS")
print("="*80)

if df_matches_2024 is not None:
    print("\n--- 2024 MATCHES ---")
    print(f"Total Matches: {len(df_matches_2024):,}")
    print(f"Total Columns: {len(df_matches_2024.columns)}")
    print(f"\nColumn Names:\n{list(df_matches_2024.columns)}")
    
    print("\n--- SAMPLE DATA (First 3 matches) ---")
    print(df_matches_2024.head(3).T)  # Transpose for better readability
    
    print("\n--- DATA TYPES ---")
    print(df_matches_2024.dtypes)
    
    print("\n--- MISSING VALUES (Top 20 columns) ---")
    missing = df_matches_2024.isnull().sum()
    missing_pct = (missing / len(df_matches_2024) * 100).round(2)
    missing_df = pd.DataFrame({
        'Missing_Count': missing,
        'Missing_Percent': missing_pct
    })
    print(missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False).head(20))
    
    # Date analysis
    if 'tourney_date' in df_matches_2024.columns:
        # Try to parse date
        df_matches_2024['tourney_date_parsed'] = pd.to_datetime(
            df_matches_2024['tourney_date'], 
            format='%Y%m%d', 
            errors='coerce'
        )
        print(f"\n--- DATE RANGE ---")
        print(f"First Match: {df_matches_2024['tourney_date_parsed'].min()}")
        print(f"Last Match: {df_matches_2024['tourney_date_parsed'].max()}")
        print(f"Date Span: {(df_matches_2024['tourney_date_parsed'].max() - df_matches_2024['tourney_date_parsed'].min()).days} days")
    
    # Tournament analysis
    if 'tourney_name' in df_matches_2024.columns:
        print(f"\n--- TOURNAMENT ANALYSIS ---")
        print(f"Unique Tournaments: {df_matches_2024['tourney_name'].nunique()}")
        print(f"\nTop 10 Tournaments by Match Count:")
        print(df_matches_2024['tourney_name'].value_counts().head(10))
    
    # Player analysis
    winner_col = [col for col in df_matches_2024.columns if 'winner' in col.lower() and 'name' in col.lower()]
    loser_col = [col for col in df_matches_2024.columns if 'loser' in col.lower() and 'name' in col.lower()]
    
    if winner_col and loser_col:
        print(f"\n--- PLAYER PARTICIPATION ---")
        winner_col = winner_col[0]
        loser_col = loser_col[0]
        
        all_players = pd.concat([
            df_matches_2024[winner_col],
            df_matches_2024[loser_col]
        ])
        print(f"Unique Players in 2024: {all_players.nunique()}")
        print(f"\nMost Active Players (by matches):")
        print(all_players.value_counts().head(10))
    
    # Score analysis
    score_col = [col for col in df_matches_2024.columns if 'score' in col.lower()]
    if score_col:
        print(f"\n--- SCORE COMPLETENESS ---")
        print(f"Matches with scores: {df_matches_2024[score_col[0]].notna().sum():,}")
        print(f"Matches without scores: {df_matches_2024[score_col[0]].isna().sum():,}")

# ============================================================================
# SECTION 5: RELATIONSHIP ANALYSIS
# ============================================================================

print("\n" + "="*80)
print("SECTION 5: RELATIONSHIP ANALYSIS")
print("="*80)

# Check if player IDs in matches exist in players table
if df_players is not None and df_matches_2024 is not None:
    print("\n--- FOREIGN KEY VALIDATION: Matches â†’ Players ---")
    
    # Find winner/loser ID columns
    winner_id_col = [col for col in df_matches_2024.columns if 'winner_id' in col.lower()]
    loser_id_col = [col for col in df_matches_2024.columns if 'loser_id' in col.lower()]
    
    if winner_id_col and loser_id_col and 'player_id' in df_players.columns:
        winner_id_col = winner_id_col[0]
        loser_id_col = loser_id_col[0]
        
        all_match_player_ids = pd.concat([
            df_matches_2024[winner_id_col],
            df_matches_2024[loser_id_col]
        ]).dropna().unique()
        
        player_ids_in_master = df_players['player_id'].unique()
        
        print(f"Unique player IDs in matches: {len(all_match_player_ids):,}")
        print(f"Unique player IDs in players table: {len(player_ids_in_master):,}")
        
        # Find orphaned IDs
        orphaned = set(all_match_player_ids) - set(player_ids_in_master)
        print(f"Player IDs in matches but NOT in players table: {len(orphaned):,}")
        
        if len(orphaned) > 0:
            print(f"Sample orphaned IDs: {list(orphaned)[:10]}")

# Check if player IDs in rankings exist in players table
if df_players is not None and df_rankings_current is not None:
    print("\n--- FOREIGN KEY VALIDATION: Rankings â†’ Players ---")
    
    if 'player' in df_rankings_current.columns and 'player_id' in df_players.columns:
        ranking_player_ids = df_rankings_current['player'].dropna().unique()
        player_ids_in_master = df_players['player_id'].unique()
        
        print(f"Unique player IDs in rankings: {len(ranking_player_ids):,}")
        print(f"Unique player IDs in players table: {len(player_ids_in_master):,}")
        
        orphaned = set(ranking_player_ids) - set(player_ids_in_master)
        print(f"Player IDs in rankings but NOT in players table: {len(orphaned):,}")
        
        if len(orphaned) > 0:
            print(f"Sample orphaned IDs: {list(orphaned)[:10]}")

# ============================================================================
# SECTION 6: DATA QUALITY ISSUES
# ============================================================================

print("\n" + "="*80)
print("SECTION 6: DATA QUALITY ISSUES SUMMARY")
print("="*80)

print("\n--- ISSUES TO ADDRESS ---")
issues = []

if df_players is not None:
    # Check for duplicates
    if 'player_id' in df_players.columns:
        dupes = len(df_players) - df_players['player_id'].nunique()
        if dupes > 0:
            issues.append(f"Players: {dupes} duplicate player_ids")
    
    # Check missing critical fields
    critical_cols = ['player_id', 'name_first', 'name_last']
    for col in critical_cols:
        if col in df_players.columns:
            missing = df_players[col].isna().sum()
            if missing > 0:
                issues.append(f"Players: {missing} missing {col}")

if df_matches_2024 is not None:
    # Check for matches without scores
    score_cols = [col for col in df_matches_2024.columns if 'score' in col.lower()]
    if score_cols:
        missing_scores = df_matches_2024[score_cols[0]].isna().sum()
        if missing_scores > 0:
            issues.append(f"Matches 2024: {missing_scores} matches without scores ({missing_scores/len(df_matches_2024)*100:.1f}%)")

if len(issues) == 0:
    print("âœ“ No major data quality issues detected!")
else:
    for i, issue in enumerate(issues, 1):
        print(f"{i}. {issue}")

print("\n" + "="*80)
print("ANALYSIS COMPLETE")
print("="*80)
```
